```python
#1.USING THE K NEAREST NEIGHBOURS
### hyper parameter tunng and building and fiiting the pipeline,handling the data imbalance
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Combine preprocessing and model in a pipeline
knn_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier(weights='distance')) # Using 'distance' weights to give more importance to closer neighbors
])

# Define the parameter grid for GridSearchCV
param_grid = {
    'knn__n_neighbors': [3, 5, 7, 9]
}

# Instantiate GridSearchCV
grid_search = GridSearchCV(knn_pipeline, param_grid, cv=5, scoring='recall') # Scoring on recall to prioritize minority class

# Fit GridSearchCV to the training data
grid_search.fit(x_processed, y_processed)

# Get the best model from the grid search
best_knn = grid_search.best_estimator_

print("Best parameters found: ", grid_search.best_params_)
```

```python
# Learning history plotting function
def plot_history(history):
  """Takes a keras model learning history and plots each metric"""

  metrics = history.history.keys()

  for metric in metrics:
      if not 'val' in metric:
        plt.plot(history.history[f'{metric}'], label=f'{metric}')
        if f'val_{metric}' in metrics:
          plt.plot(history.history[f'val_{metric}'], label=f'val_{metric}')
        plt.legend()
        plt.title(f'{metric}')
        plt.show()
```

Let's break down the learning curves you've plotted for your deep learning model:

**1. Accuracy Curve:**

*   **Training Accuracy (accuracy):** This curve shows how well your model is performing on the training data as the training progresses through each epoch. A rising accuracy generally indicates that the model is learning the patterns in the training data.
*   **Validation Accuracy (val_accuracy):** This curve shows how well your model is performing on the unseen test data. This is a crucial metric for evaluating if your model is generalizing well and not just memorizing the training data.

**Interpretation:**

*   If the training accuracy is much higher than the validation accuracy, it suggests that the model is overfitting to the training data.
*   If both curves are low, it might indicate underfitting (the model is not complex enough to capture the patterns).
*   Ideally, you want both curves to increase and converge, indicating good learning and generalization.

**2. Loss Curve:**

*   **Training Loss (loss):** This curve shows the error of your model on the training data. A decreasing loss indicates that the model is improving its predictions on the training data.
*   **Validation Loss (val_loss):** This curve shows the error of your model on the unseen test data. This is another important indicator of generalization.

**Interpretation:**

*   Similar to accuracy, if the training loss is significantly lower than the validation loss, it suggests overfitting.
*   If both losses are high, it might indicate underfitting.
*   Ideally, both curves should decrease and converge to a low value.

**3. Precision Curve:**

*   **Training Precision (precision\_2):** This metric shows the precision of your model on the training data. Precision is the ratio of true positive predictions to the total number of positive predictions (true positives + false positives). In your case, it's the proportion of loans predicted as "Charged Off" that were actually "Charged Off".
*   **Validation Precision (val\_precision\_2):** This shows the precision on the test data.

**Interpretation:**

*   A low precision for the minority class (Class 0, "Charged Off") indicates that the model is making a lot of false positive predictions, meaning it's incorrectly flagging many "Fully Paid" loans as "Charged Off".

**4. Recall Curve:**

*   **Training Recall (recall\_2):** This metric shows the recall of your model on the training data. Recall is the ratio of true positive predictions to the total number of actual positives (true positives + false negatives). In your case, it's the proportion of actual "Charged Off" loans that your model correctly identified.
*   **Validation Recall (val\_recall\_2):** This shows the recall on the test data.

**Interpretation:**

*   A low recall for the minority class ("Charged Off") indicates that the model is making a lot of false negative predictions, meaning it's failing to identify many of the actual "Charged Off" loans. This is a significant issue in a loan eligibility context, as it means the model is approving loans that are likely to default.

**Overall Interpretation of Your Curves:**

Looking at your specific plots, the validation recall quickly jumps to 1.0, while the validation precision remains around 0.7. The validation accuracy also plateaus around 0.7. This, combined with the classification reports you saw earlier, strongly suggests that your deep learning model is also primarily predicting the majority class ("Fully Paid"). A recall of 1.0 for the minority class on the validation set would mean the model correctly identified *all* of the "Charged Off" loans, which is highly unlikely given the class imbalance and the other metrics. It's more probable that the model is predicting "Fully Paid" for almost all instances, resulting in a high recall for the majority class and a misleadingly high recall for the minority class in the early epochs due to random chance, before settling into predicting only the majority class.

This reinforces the challenge of training models on imbalanced datasets, where models tend to favor the majority class. You'll need to implement techniques to address this imbalance during training to improve the model's ability to correctly identify the "Charged Off" loans.
